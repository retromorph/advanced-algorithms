# Введение в алгоритмы

## Классификация алгоритмов

По характеру решения:
1. Точные алгоритмы - алгоритмы для которых доказано, что они выдают наилучшее решение
2. Приближенные алгоритмы - алгоритмы, для которых известна величина $|C - C'| < c$, где $C$ - оптимальное решение, $C'$ - приближенное решение
3. Эвристические алгоритмы - алгоритмы, которые почему-то работают на практике, но мы точно не знаем почему.
4. Вероятностные алгоритмы - алгоритмы, для которых задана точность $r < 1$

По порядку выполнения:
1. Последовательные алгоритмы
2. Параллельные алгоритмы

По определенности:
1. Детерминированные алгоритмы - заранее известен путь, по которому будет получено решение
2. Рандомизированные алгоритмы - соответственно, путь заранее не известен

**Определение:** Задача принятия решения - Вход $\rightarrow$ алгоритм $\rightarrow$ $\{True, False\}$
_Пример:_ Проверка графа на связность

**Определение:** Сертификат решения - набор данных, позволяющий оценить корректность работы алгоритма
_Пример:_ При проверке наличия пути между вершинами $u$ и $v$ в графе, сертификатом будет являться сам путь

**Определение:** Задача оптимизации - так подобрать степени свободы, чтобы получить максимальное или минимальное значение функции в какой-либо метрике

**Определение:** Offline алгоритм - алгоритм, все входные данные которого заранее известны
**Определение:** Online алгоритм - интерактивный алгоритм

## RAM-модель

**Определение:** RAM - random access machine. В рамках этой модели:
1. Память одноранговая бесконечная с произвольным доступом. Индексируется с $0$
2. Для исполнения любой простой операции (арифметической, логической, битовой и функциональной) требуется ровно один временной шаг
3. Циклы и подпрограммы не считаются простыми операциями

Понятно, что RAM модель имеет посредственное отношение к том, как в действительности работает компьютер, но это удобная и достаточно устойчивая теоретическая конструкция, позволяющая оценивать время работы программы машинно-независимым путем. Достаточно сложно найти алгоритм, существенно нарушающий допущения модели RAM.

Рассмотрим некоторые теоретические надстройки над RAM:
1. Ограничение по памяти $m(n) = O(t(n))$
2. Ограничение по размеру числа
3. Параллельным алгоритмам может потребоваться разделение по памяти, его можно организовать с помощью деления по модулю
4. Генерация случайных чисел $randReal \rightarrow [0, 1]$
