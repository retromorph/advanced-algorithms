# Декартово Дерево

## Введение

Декартово дерево - дерево с двумя ключами в каждой вершине. По первому ключу декартово дерево - бинарное дерево поиска, по второму - бинарной кучей

```c++
#include <bits/stdc++.h>
using namespace std;

// Treap (Cartesian Tree) implementation in C++
// Supports insert, erase, find, split, merge, and order statistics

struct Treap {
    struct Node {
        int key, priority;
        int size;
        Node *l, *r;
        Node(int k)
            : key(k), priority(rand()), size(1), l(nullptr), r(nullptr) {}
    };

    using pNode = Node*;
    pNode root = nullptr;

    // Update subtree size
    void upd_size(pNode t) {
        if (t) {
            t->size = 1;
            if (t->l) t->size += t->l->size;
            if (t->r) t->size += t->r->size;
        }
    }

    // Split treap t into [< key] and [>= key]
    void split(pNode t, int key, pNode &l, pNode &r) {
        if (!t) { l = r = nullptr; }
        else if (key < t->key) {
            split(t->l, key, l, t->l);
            r = t;
            upd_size(r);
        } else {
            split(t->r, key, t->r, r);
            l = t;
            upd_size(l);
        }
    }

    // Merge two treaps l and r, all keys in l <= keys in r
    pNode merge(pNode l, pNode r) {
        if (!l || !r) return l ? l : r;
        if (l->priority > r->priority) {
            l->r = merge(l->r, r);
            upd_size(l);
            return l;
        } else {
            r->l = merge(l, r->l);
            upd_size(r);
            return r;
        }
    }

    // Insert key into treap (no duplicates)
    void insert(int key) {
        if (find(key)) return;
        pNode n = new Node(key);
        pNode t1, t2;
        split(root, key, t1, t2);
        root = merge(merge(t1, n), t2);
    }

    // Erase key from treap
    void erase(int key) {
        pNode t1, t2, t3;
        split(root, key, t1, t2);
        split(t2, key + 1, t2, t3);
        // t2 contains nodes equal to key
        delete_subtree(t2);
        root = merge(t1, t3);
    }

    // Find key exists
    bool find(int key) {
        pNode t = root;
        while (t) {
            if (key == t->key) return true;
            t = (key < t->key ? t->l : t->r);
        }
        return false;
    }

    // K-th smallest (0-indexed), returns INT_MIN if out of bounds
    int kth(int k) {
        pNode t = root;
        while (t) {
            int lsz = (t->l ? t->l->size : 0);
            if (k < lsz) t = t->l;
            else if (k == lsz) return t->key;
            else { k -= lsz + 1; t = t->r; }
        }
        return INT_MIN; // invalid
    }

    // Number of keys < key
    int count_less(int key) {
        pNode t = root;
        int cnt = 0;
        while (t) {
            if (t->key < key) {
                cnt += 1 + (t->l ? t->l->size : 0);
                t = t->r;
            } else t = t->l;
        }
        return cnt;
    }

    // Cleanup entire treap
    void clear() { delete_subtree(root); root = nullptr; }

private:
    void delete_subtree(pNode t) {
        if (!t) return;
        delete_subtree(t->l);
        delete_subtree(t->r);
        delete t;
    }
};

// Example usage:
int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    Treap t;
    t.insert(5);
    t.insert(2);
    t.insert(8);
    cout << "Find 2? " << (t.find(2) ? "Yes" : "No") << "\n";
    cout << "1st smallest (0-index): " << t.kth(1) << "\n";
    cout << "Count less than 6: " << t.count_less(6) << "\n";
    t.erase(2);
    cout << "Find 2 after erase? " << (t.find(2) ? "Yes" : "No") << "\n";

    return 0;
}
```
